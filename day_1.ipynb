# Match & Subtract

```python
sorted_ranks = np.sort(left_rank)
difference_list = []

for rank in sorted_ranks:
    print(f"Current Rank: {int(rank)}/{len(sorted_ranks)}")
    # Identify indices of elements from left and right lists with the specified rank
    i_left = np.where(left_rank == rank)[0][0]
    i_right = np.where(right_rank == rank)[0][0]
    # print("Indices: ", i_left, i_right)

    # Get corresponding elements in left and right lists
    element_left = df_lists.iloc[i_left]["left"]
    element_right = df_lists.iloc[i_right]["right"]
    # print("Elements", element_left, element_right)

    # Get difference between the elements from left and right lists
    d = np.abs(element_right - element_left)
    # print(f"Difference: {d}")
    difference_list.append(d)

    # print("-----------")
```

# Summing

```python
plt.hist(difference_list, bins=50)
plt.xlabel("Differences")
plt.ylabel("Frequency")
plt.show()
```

```python
# Sum of differences
total_sum = np.sum(difference_list)
print(f"Sum of differences: {int(total_sum)}")
```
